```python
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
from langchain_community.document_loaders import PyPDFLoader
load_dotenv()
```

```python
loader = PyPDFLoader("tasks_01.pdf")

docs = loader.load()
```

```python
import os
```

```python
embeddings = OpenAIEmbeddings(
    model="thenlper/gte-base", # Make sure this is in your .env file
    base_url="https://openrouter.ai/api/v1",
    default_headers = {
        "HTTP-Referer": "http://localhost",
        "X-Title": "LangChain Learning Project"
    }
)
```

```python
vector_store = Chroma(
    embedding_function = embeddings,
    persist_directory= "My_chroma_db",
    collection_name= "tasks_01_collection"
)
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=["embeddings", "metadatas", "documents"])
```
